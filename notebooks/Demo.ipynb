```python
import pickle
import pandas as pd
from sklearn.datasets import load_breast_cancer

# load the saved model
with open('../models/log_reg_breast_cancer.pkl', 'rb') as f:
    model = pickle.load(f)

# load data
data = load_breast_cancer(as_frame=True)
df = data.frame
X = df.drop("target", axis=1)
y = df["target"]

# run a prediction on the first 5 samples
sample = X.iloc[:5]
predictions = model.predict(sample)
actual = y.iloc[:5].values

for i, (pred, actual) in enumerate(zip(predictions, actual)):
    pred_label = data.target_names[pred]
    actual_label = data.target_names[actual]
    print(f"Sample {i+1} | Predicted: {pred_label} | Actual: {actual_label}")
```
